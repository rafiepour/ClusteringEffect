```python
results = list()
#nodes=[500,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,20000,30000,40000,50000,60000,70000,80000,90000,100000,200000,300000,400000,500000,600000,700000,800000,900000,1000000,1100000,1200000,1300000,1400000,1500000]
#nodes=[100000,]
nodes=[200000,300000,400000,500000,600000,700000,800000,900000,1000000,1100000,1200000,1300000,1400000,1500000,1600000,1700000,1800000,1900000,2000000,]
#nodes=[500,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,20000,30000,40000,50000,60000,70000,80000,90000,100000]
```

```python
def job(n_count,probs,thread_results,id):
    g = nx.generators.random_graphs.barabasi_albert_graph(n=n_count,m=1)
    new_g = rm.rafie_graph(g,d=probs)
    tl = nx.average_clustering(new_g)
    jac = acu.pred_jac(new_g)
    het = acu.pred_hetero(new_g)
    thread_results[id]=((probs,tl,jac,het))
    return tl,jac,het
```

```python
try_num=10
#global thread_results
for n_count in nodes:
    step=10
    i=110.0
    n_results=list()
    th=[]
    while i<1000:
        probs = i/1000.0
        i+=step
        manager = multiprocessing.Manager()
        return_dict = manager.dict()
        for o in range(0,try_num):
            th1 = multiprocessing.Process(target=job,args=(n_count,probs,return_dict,o))
            th.append(th1)
            th1.start()
        for t in th:
            t.join()
            #tl,jac,het = job(n_count,probs)
        #    print(probs)
        #tl,jac,het = job(n_count,probs)

        pr=0.0
        tl=0.0
        jac=0.0
        het=0.0
        #print (len(return_dict))
        for tr in return_dict.values():
            pr+=float(tr[0])/float(try_num)
            tl+=(float(tr[1])/float(try_num))
            jac+=(float(tr[2])/float(try_num))
            het+=(float(tr[3])/float(try_num))
        th.clear()
        n_results.append((probs,tl,jac,het))

        if (het<=jac):
            print ("--------------------")
            print ("N:")
            print (n_count)
            print ("p:")
            print (pr)
            print ("average clustering:")
            print (tl)
            print ("jaccard:")
            print (jac)
            print ("hetero:")
            print (het)
            break
        
    total_results.append((n_count,n_results))
```
